# Makine Öğrenmesi Nedir? 

Makine Öğrenimi, adından da anlaşılacağı gibi, makinelere açık bir şekilde programlama yapmadan, belirli bir veri kümesindeki deneyimleri, gözlemleri ve modelleri analiz ederek bağımsız bir şekilde öğrenme yeteneği sağlar. Belirli bir amaç için bir program veya kod yazdığımızda, aslında makinenin izleyeceği belirli bir talimatlar dizisi yazıyoruz.

Makine öğreniminde ise, makinenin veri setindeki örüntüleri tanımlayıp analiz ederek öğreneceği bir algoritma kullanıyoruz. Makinelere veri setinden gözlemlerine ve öğrendiklerine dayanarak bağımsız olarak kararlar almayı öğreneceği bir yetenek kazandırıyoruz.

Bir fotoğraf yüklediğinizde sizden fotoğrafları etiketlemenizi isteyen Facebook’un yüz tanıma algoritmasını düşünün. Sesli asistanlar bile kullanıcının isteğini belirlemek ve hizmet vermek için makine öğrenimini kullanır. Tesla’nın otopilot özelliği başka bir örnektir.

# Makine Öğrenmesinin Geçmişi

Makine öğrenimi yeni bir kavram değil. Aslında, sinir ağları ilk olarak 1943 yılında bir araştırma makalesinde yeni bir kavram olarak tanıtıldı.

İlk günlerde makine öğrenimindeki ilerleme, bu alanı yalnızca büyük akademik kurumlar veya çok uluslu şirketler için erişilebilir hale getiren yüksek bilgi işlem maliyeti nedeniyle biraz yavaş olmuştur. Bir şirketin ihtiyaçları için verilerin kendi içinde elde edilmesinin zor olduğu gerçeği de vardı. Ancak internetin gelişiyle birlikte artık her gün kentilyonlarca veri üretiyoruz. Bunu hesaplama maliyetlerinin düşüşü ile birleştirin.

**1943:** Makine öğrenimi tarihi, Walter Pitts ve Warren McCulloch tarafından yazılan "Sinirsel aktivitede içkin olan fikirlerin mantıksal hesabı" adlı bilimsel makalede sunulan sinir ağlarının ilk matematiksel modeliyle başlar.

<img src='https://historyofinformation.com/images/Screen_Shot_2020-09-09_at_6.46.46_AM_thumb.png'></img>


**1949:** Donald Hebb'in "The Organization of Behavior" adlı kitabı yayınlandı. Bu kitapta, davranışın sinir ağları ve beyin aktivitesiyle nasıl ilişkili olduğuna dair teoriler vardır ve makine öğrenimi gelişiminin önemli kaynaklarından birisi olmuştur.

<img src='https://upload.wikimedia.org/wikipedia/en/0/08/The_Organization_of_Behavior.jpg'></img>

**1950:** Makine öğreniminde öncü olan Arthur Samuel, şampiyona düzeyinde dama oyunlarını bilgisayarda oynamak için bir program oluşturdu. Oyun, her olası yolu araştırmak yerine, kazanma şansını ölçen alfa-beta testlerini kullandı. Ek olarak, Samuel, rakibin de en iyi şekilde oynadığını varsayarak, en uygun hamleyi bulmak için bir minimax algoritmasını (bugün hala yaygın olarak kullanılan) oluşturdu. Ayrıca, örneğin önceki pul hareketlerini hatırlayarak ve bunları kazanma şansı ile karşılaştırarak, programının sürekli olarak iyileştirilmesi için mekanizmalar tasarladı. Arthur Samuel, "makine öğrenimi" terimini bulan ve popüler hale getiren ilk kişidir.

<img src='https://history-computer.com/ModernComputer/thinkers/images/Arthur-Samuel2.jpg'></img>


**1951:** Çoğu bilgisayar çalıştırmak için delikli kartları kullanmaya devam ederken, Marvin Minsky ve Dean Edmonds, kısa ve uzun süreli belleğe sahip 40 adet birbirine bağlı nörondan oluşan ilk yapay sinir ağını kurdu.
<img src='http://cyberneticzoo.com/wp-content/uploads/neuron-SNARC-GJLoan2011-x640.jpg'></img>


**1956:** Dartmouth Çalıştayı'na "yapay zekanın doğduğu yer" denilir. İki aylık bir süre boyunca, matematik, mühendislik, bilgisayar ve bilişsel bilimler alanlarında önde gelen bir grup bilim insanı, yapay zeka ve makine öğrenmesi araştırma alanlarını oluşturmak ve beyin fırtınası yapmak için bir araya geldi.

<img src='https://miro.medium.com/max/620/0*3jGv1es40NHPcTE3.jpg'></img>

**1965:** Ukrayna doğumlu sovyet bilim adamları Alexey (Oleksii) Ivakhnenko ve Valentin Lapa, polinom aktivasyon işlevini kullanan ve Grup Veri İşleme Yöntemi (GMDH) kullanılarak eğitilen sinir ağının hiyerarşik temsilini geliştirdiler. Bu çalışma, ilk çok katmanlı algılayıcı olarak kabul edilir ve Ivakhnenko genellikle derin öğrenmenin babası olarak kabul edilir.

**1967:** Stanford'dan Thomas Cover ve Peter E. Hart,  IEEE Transactions on Information Theory (Volume: 13, Issue: 1, January 1967, pages 22-27) dergisinde en yakın komşu algoritması (makine öğreniminde sınıflandırma ve regresyon için kullanılır) hakkında bir makale yayınladılar.

**1997:** IBM satranç bilgisayarı Deep Blue, satrançta dünya şampiyonu Garry Kasparov'u yendi. O zamanlar bu başarı, insan zekasını yakalayan makinelerin bir kanıtı olarak görülüyordu.

<img src='https://media.pri.org/s3fs-public/styles/story_main/public/story/images/DeepBlue.png?itok=N01yMnCK'></img>

**2000:** Ukrayna doğumlu sinir ağları araştırmacısı Igor Aizenberg çalışmalarında, "derin öğrenme" teriminden ilk kez söz etti.

**2009:** ImageNet oluşumu, Fei-Fei Li tarafından başlatıldı. Li, yapay zeka ve makine öğreniminin gerçekten pratik ve kullanışlı olabilmeleri için gerçek dünyayı yansıtan iyi eğitim verilerine sahip olması gerektiğine inandığından, algoritmaları eğitmek için mevcut verileri genişletmek istedi. The Economist dergisi, bu veritabanının oluşturulmasını, yapay zekayı tüm teknoloji topluluğu genelinde yaygınlaştırmak için olağanüstü bir olay olarak tanımladı.


<img src='https://devopedia.org/images/article/172/7316.1561043304.png'></img>


**2012:** Kapsamlı bir makine öğrenimi geçmişine sahip olan Google'ın X Lab ekibi,  görüntü işlemede ünlü hale gelen ve görüntülerdeki kedileri tanımlayabilen yapay zeka algoritmasını geliştirdi.

**2014:** Facebook araştırma ekibi, derin öğrenmeyi kullanan bir yüz tanıma sistemi olan DeepFace'i geliştirdi. Facebook kullanıcılarının 4 milyon görüntüsü üzerinde eğitilen dokuz katmanlı sinir ağı, görüntülerdeki insan yüzlerini insanlarla aynı doğrulukla tespit edebiliyordu (yaklaşık % 97.35).

**2015:** DeepMid Technologiesşirketi tarafından geliştirilen AlphaGo programı, profesyonel bir Go oyuncusunu yenen ilk yapay zekadır. Go, daha önce bir bilgisayara öğretmek için neredeyse imkansız bir oyun olduğu düşünülen en eski ve en zor soyut strateji oyunlarından biridir.

**2016:** Bir grup bilim insanı, Bilgisayarla Görme ve Örüntü Tanıma Konferansı sırasında Face2Face'i sundu. Mantığı ve algoritmaları, günümüzün "deepfake" yazılımının çoğunun temelini oluşturur.

# Karışıklığı Giderme: Yapay Zeka - Makine Öğrenimi - Derin Öğrenme Farkları

<img src='https://miro.medium.com/max/631/0*Q3PICBlib-932hhH.png'></img>

Adından da anlaşılacağı gibi **yapay zeka** (AI), insan zekasını makinelere dahil etmek anlamına gelecek şekilde kaba bir şekilde yorumlanabilir. Bir makine, sorunları çözen bir dizi öngörülen kurala dayalı olarak (algoritmalar) görevleri tamamladığında, bu tür bir "akıllı" davranış yapay zeka olarak adlandırılır.

Örneğin, bu tür makineler nesneleri hareket ettirebilir ve manipüle edebilir, birinin ellerini kaldırıp kaldırmadığını fark edebilir veya başka sorunları çözebilir.

Yapay zeka destekli makineler genellikle iki gruba ayrılır - genel ve dar kapsamlı makineler:

Genel kapsamlı yapay zeka makineleri, yukarıda belirtilenler gibi sorunları akıllıca çözebilir. Dar kapsamlı yapay zeka makineleri, kapsamları sınırlı olsa da, belirli görevleri çok iyi, bazen insanlardan daha iyi gerçekleştirebilir.
Pinterest'teki resimleri sınıflandırmak için kullanılan teknoloji, dar kapsamlı bir yapay zeka örneğidir.

Adından da anlaşılacağı gibi, **makine öğrenmesi** (ML), bilgisayar sistemlerini "öğrenme" yeteneği ile güçlendirmek anlamına gelecek şekilde yorumlanabilir.
Makine öğreniminin amacı, makinelerin sağlanan verileri kullanarak kendi kendilerine öğrenmelerini ve doğru tahminlerde bulunmalarını sağlamaktır.
Makine öğrenimi, yapay zekanın bir alt kümesidir; aslında, bu sadece yapay zekayı gerçekleştirmek için kullanılan bir tekniktir.

Makine öğrenmesi, algoritmaları nasıl karar vereceklerini öğrenebilecekleri şekilde eğitme yöntemidir. Makine öğreniminde eğitim, algoritmaya çok fazla veri vermeyi ve işlenen bilgiler hakkında daha fazla bilgi edinmesine izin vermeyi gerektirir. Örneğin, özelliklerine göre meyvenin türünü tanımlayan bir tablo:

>Kütle (Gr) | Yüzey | Meyve Türü
>--- | --- | ---
>155 | Pürüzlü | Portakal
>180 | Pürüzlü | Portakal
>135 | Düz | Elma
>110 | Düz | Elma
>120 | Düz| **?**

Yukarıdaki tablodan da görebileceğiniz gibi meyveler kütlelerine ve yüzey dokularına göre farklılaşmaktadır. Son sıradaki meyvenin türünü bilmemekteyiz.
Bu meyvenin portakal mı yoksa elma mı olduğunu belirlemeye çalışmak için bir makine öğrenimi algoritması geliştirilebilir.

Algoritma eğitim verileriyle beslendikten sonra, bir portakal ve bir elma arasındaki farklı özellikleri öğrenecektir. Kütle ve yüzey verileri algoritmaya doğru bir şekilde sunulursa, algoritma bu özelliklere sahip meyve türünü doğru bir şekilde tahmin edebilir.

Daha önce belirtildiği gibi, **derin öğrenme** (DÖ) makine öğreniminin bir alt kümesidir; aslında, makine öğrenimini gerçekleştirmeye yönelik bir tekniktir. Başka bir deyişle, DÖ, makine öğreniminin bir sonraki evrimidir.
DL algoritmaları, kabaca insan beyninde bulunan bilgi işleme modellerinden esinlenmiştir.

Tıpkı örüntüleri belirlemek ve çeşitli bilgi türlerini sınıflandırmak için beynimizi kullandığımız gibi, derin öğrenme algoritmalarının makineler için aynı görevleri yerine getirmesi öğretilebilir.Beyin genellikle aldığı bilgileri deşifre etmeye çalışır. Bunu, öğeleri etiketleyerek ve çeşitli kategorilere ayırarak gerçekleştirir.

Yeni bir bilgi aldığımızda, beyin onu anlamadan önce bilinen bir öğeyle karşılaştırmaya çalışır - bu, derin öğrenme algoritmalarının kullandığı yöntemle aynıdır.

Örneğin, yapay sinir ağları (ANN), beynimizin karar alma şeklini taklit etmeyi amaçlayan bir tür algoritmadır. Derin öğrenme ile makine öğrenimini karşılaştırmak, onların ince farklarını anlamamıza yardımcı olabilir.
Örneğin, DÖ sınıflandırma için kullanılacak özellikleri otomatik olarak keşfedebilirken, MÖ bu özelliklerin manuel olarak sağlanmasını gerektirir.
Dahası, MÖ'nin aksine, DÖ doğru sonuçlar elde etmek için üst düzey makinelere ve önemli miktarda eğitim verisine ihtiyaç duyar.

# Makine Öğrenmesinin Bileşenleri
## Verinin Toplanması ve Hazırlanması

Makine öğreniminin temellerindeki ilk adım, makineye bilgi / veri beslememizdir. Bu veriler iki bölüme ayrılır: eğitim verileri ve test verileri.

Bir kişinin fotoğrafı gösterilir göstermez tanımlayabilen bir yazılım geliştirmek istediğimizi düşünün. Veri toplayarak başlıyoruz, yani insanların fotoğraflarını ediniyoruz. Şimdi bu aşamada, verilerimizin tüm popülasyonu temsil ettiğinden emin olmalıyız, yani, sadece 20-40 yaş arası yetişkinleri dahil edersek, bir bebeğin resmi gösterilirse yazılım başarısız olacaktır.

Veriler, bir kez yeterince eğitilmiş modelin daha sonra test edilebilmesini sağlamak için genellikle 80/20 veya 70/30 olarak bölünür.

<img src='https://image.slidesharecdn.com/mlregression-splittingdatasets-170914113926/95/machine-learning-splitting-datasets-3-638.jpg?cb=1505389251'></img>


## Modelin Seçilmesi ve Eğitilmesi

Bu, makine öğrenimiyle ilgili temel işlemlerdeki ikinci adımdır. Belirli bir problem türünü çözebilmek için daha da geliştirilmiş ve değiştirilmiş çeşitli makine öğrenimi algoritmalarımız ve modellerimiz bulunmaktadır. Bu nedenle, elimizdeki soruna uygunluğuna bağlı olarak bir model seçip eğitmemiz zorunludur.

<img src='https://docs.microsoft.com/en-us/azure/machine-learning/media/algorithm-cheat-sheet/machine-learning-algorithm-cheat-sheet.svg'></img>


## Modelin Değerlendirilmesi

Makine, örüntüleri ve özellikleri eğitim verilerinden öğrenir, ve yeni verileri tanımlama, sınıflandırma veya tahmin etme gibi kararlar almak için kendini eğitir. Makinenin bu kararları ne kadar doğru bir şekilde alabildiğini kontrol etmek için tahminler test verileri üzerinde değerlendirilir.

Bu durumda, önce eğitim verileri üzerinde çalışacağız ve model yeterince eğitildikten sonra, fotoğraftaki yüzleri tanımada ne kadar başarılı olduğunu anlamak için test verileri üzerinde kullanacağız.


## Hiperparametreler ve Tahmin

Makine öğrenimi terminolojisinde hiperparametreler, modelin kendisi tarafından tahmin edilemeyen parametrelerdir. Ancak modelin performansını artırmada çok önemli bir rol oynadıkları için yine de bunları hesaba katmamız gerekir.

Genel konuşursak, bir makine öğrenme modelindeki hiperparametreler, algoritmayı çalıştırmak için kullanıcı tarafından belirtilmesi gereken parametrelerdir. Klasik parametreler verilerden öğrenilir, hiperparametreler verilerden öğrenilebilir veya öğrenilmeyebilir.

<img src='https://miro.medium.com/max/700/0*KKjR5PliXjgLrX8X.png'></img>

## Performans-Görev-Deneyim

Bir makine öğrenmesi çalışması üç bileşene sahiptir: **P, T, E**

P performance(performans), T task(görev) ve E experience(deneyim)

Arthur Samuel’in üzerinde çalıştığı dama oyununu örnek verelim. E = birçok dama oyununu oynama deneyimi(experience), T = dama oynama görevi(task) ve P = programın bir sonraki oyunu kazanma ihtimali (performansı) olarak gösterebiliriz.

# Python'da Makine Öğrenmesi Kütüphaneleri

## Scikit-learn

SciPy kütüphanesi üzerine kurulmuş bir Python makine öğrenmesi kütüphanesidir ve sınıflandırma, kümeleme ve regresyon dahil olmak üzere çeşitli algoritmalar içinde barındırır. Bilimsel ve sayısal hesaplamalar için NumPy ve SciPy gibi diğer Python kütüphaneleri ile birlikte kullanılabilir. Sınıflarından ve fonksiyonların bazıları sklearn.cluster, sklearn.datasets, sklearn.ensemble, sklearn.mixture vb. olarak sıralanabilir. 


## TensorFlow

TensorFlow, yüksek performanslı sayısal hesaplamalar ve sinir ağları gibi makine öğrenimi uygulamaları için açık kaynaklı bir yazılım kütüphanesidir. Esnek mimarisi sayesinde CPU'lar, GPU'lar, TPU'lar vb. gibi çeşitli platformlarda hesaplamanın kolay dağıtımına izin verir.

## Keras

Keras, sinir ağları ve diğer derin öğrenme modellerini geliştirmek için kullanılan derin öğrenme kütüphanesidir. TensorFlow, Microsoft Cognitive Toolkit veya Theano üzerine kurulabilir, modüler ve genişletilebilir olmaya odaklanır.
