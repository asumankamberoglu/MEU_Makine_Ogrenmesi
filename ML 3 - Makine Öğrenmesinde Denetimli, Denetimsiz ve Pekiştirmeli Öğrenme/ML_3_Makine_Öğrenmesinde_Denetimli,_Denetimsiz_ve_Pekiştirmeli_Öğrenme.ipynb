Makine Öğrenmesi hakkında bir şeyler duymuş olsanız da, kullanılan ana yöntemlere dair genel bir bakışa da sahip olmak gerekir. Denetimli ve Denetimsiz öğrenme, yapay zekanın (AI) makine öğrenimi alt kümesine giren iki ana alt kategoridir.

Denetim sözcüğü yerine, Türkçe literatürde kontrol, güdüm ya da gözetim çevirileri de kullanılmaktadır. 

# Denetimli Öğrenme (Supervised Learning)

Denetimli öğrenme, aldığı girdi ve çıktı verilere dayalı olarak öngörüde bulunan bir model geliştirir. Makine, bilinen girdilerden ve bilinen yanıtlardan öğrenmeye başlar. Başka bir deyişle, verisetinin sahip olduğu mevcut gerçekliğini (ground truth) kullanan bir öğrenme yöntemidir. Cevabın verilebilmesi için, örneklerden gelen bilgi kullanılır. Bu örneklere **eğitim verisi** denir.  Denetimli öğrenme, algoritmanın bu eğitim örneklerinden öğrenmesine ve doğru etiketi tahmin etmek için bir tahmin modeli kullanarak, yeni girdi verildiğinde öğrendiklerinden yola çıkarak yeni girdi veriye ilişkin etiketi tahmin etmesine yardımcı olur. Bu makine öğrenimi yöntemi, bilinen veriler için yaklaşık bir tahmin sonucu oluşturmanız gerektiğinde kullanışlıdır. Bu tür bir tahminin yararlı olduğu iki ana alan, sınıflandırma ve regresyondur.

## Sınıflandırma (Classification)

Sınıflandırma, bir algoritmanın girdi verilerin hangi kategoriye ait olduğuna karar vermesi gerektiğinde kullanılan denetimli bir öğrenme tekniğidir. Bu tekniğin yaygın olarak kullanıldığı bir örnek, elektronik postanın spam veya gerçek olup olmadığını belirleyen programlardır. Bilgisayar, sınıflandırmayı kullanarak elektronik postanın hangi sınıfa ait olduğuna karar verebilir ve onu sınıflandırabilir.

>![E-mail Spam](https://radimrehurek.com/data_science_python/plot_ML_flow_chart_11.png)


Sınıflandırma problemleri, algoritmadan ayrı bir değeri tahmin etmesini, girdi verilerini belirli bir sınıfın veya grubun üyesi olarak tanımlamasını ister. Hayvan resimlerinden oluşan bir eğitim veri setinde bu, her fotoğrafın kedi, koala veya kaplumbağa olarak önceden etiketlendiği anlamına gelir. Algoritma daha sonra diğer koalaların ve kaplumbağaların yeni görüntülerini ne kadar doğru bir şekilde sınıflandırabildiğiyle değerlendirilir.

>![Animal Recognition](https://mlwhiz.com/images/id1.png)

Denetimli öğrenme tekniklerini uydu görüntüsü sınıflandırma için de kullanabiliriz. Algoritmaya öğretilecek olan arazi örtüsü / arazi kullanımı eğitim veri setleri ile, görüntünün sınıflandırılmamış piksellerinin hangi arazi örtüsü / arazi kullanımı sınıfına denk geleceğini belirleme işlemine uydu görüntüsü sınıflandırması denir. 



>![Satellite Imagery Classification](https://seethedatablog.files.wordpress.com/2019/04/caret-experiment-rf-svm-nnet.png)

## Regresyon (Regression) 

Regresyon, veri setini sınıflara ayırmayı hedeflemez. Girdi veri üzerinden çıktı verilerin sürekli olarak tahmin edilmesini kapsar. Bilgisayar, birçok farklı regresyon algoritması tekniğini kullanarak, çıktılarındaki değişiklikler ve dalgalanmalar ile verileri tahmin edebilir. Regresyon, örneğin tıp alanında, önceki hastalardan topladıkları verileri kullanarak kalp krizlerini tahmin etmek için yardımcı olur.

Regresyon problemleri sürekli veriler ile ilgienir. Doğrusal regresyon, cebir derslerinden tanıdık gelmelidir: belirli bir x değeri verildiğinde, y değişkeninin beklenen değeri nedir?

# Denetimsiz Öğrenme (Unsupervised Learning)
Denetimsiz öğrenme, algoritmaya ilk olarak ne yapılacağını bildirmeden ve nasıl bir sonuç isteneceği tanımlanmadan etiketlenmiş örnek verilerin sunulduğu makine öğrenmesi tekniğidir. Algoritma daha sonra yapılandırılmamış veri setindeki yapısal özellikleri türeterek bulmaya çalışır. Bununla birlikte, denetimli öğrenmenin aksine, bu tekniğin sınıflandırılmış bir çıktısı yoktur. Bunun yerine, kümeleme (clustering) adı verilen başka bir teknik kullanır.

Eğitim veri kümesi, istenen belirli bir sonuç veya doğru cevaba sahip olmayan bir örnek koleksiyonudur. Sinir ağı daha sonra yararlı özellikleri çıkararak ve yapısını analiz ederek verilerdeki yapıyı otomatik olarak bulmaya çalışır.

## Kümeleme (Clustering)

Kümeleme, veri seti içinde saklı benzerlikleri bulur ve girdi örnekleri kümeler halinde gruplandırır. Bu teknik, seçmenlerin siyasi yönelimlerini belirleyen başkanlık kampanyalarında ve pazarlama faaliyetleri için kullanışlıdır. Ayrıca, girdi örnekler herhangi bir kümeye ait olmadığında, olağandışı örüntüleri bize gösterir. Örneğin bir kredi kartı aynı anda farklı yerlerde kullanılıyorsa, bankalar tarafından dolandırıcılığı tespit etmek için kullanılabilir.

## Boyut Azaltma (Dimensionality Reduction)

Boyut azaltma, tek tek özellikler arasındaki ilişkileri öğrenerek işlem yoğunluğunu azaltmak için gereksiz özellikleri ortadan kaldırma ve verileri yalnızca birbiriyle ilişkisi olan özellikleri kullanarak temsil etme yöntemidir. Bu yöntem sonucunda, başladığımızdan çok daha az özellik elde edilir.

# Pekiştirmeli Öğrenme (Reinforcement Learning)
Video oyunları pekiştirici ipuçları ile doludur. Bir seviyeyi tamamlayın ve bir rozet kazanın. Kötü adamı belirli sayıda hamlede yenin ve bonus kazanın. Bir tuzağa adım atın - oyun biter!

Bu ipuçları, oyuncuların bir sonraki oyun için performanslarını nasıl geliştireceklerini öğrenmelerine yardımcı olur. Bu geri bildirim olmasaydı, oyuncular bir sonraki seviyeye geçme umuduyla bir oyun ortamında rastgele eylemlerde bulunacaklardı.

Pekiştirmeli öğrenme (Reinforcement Learning) aynı prensip üzerinde çalışır - ve aslında video oyunları bu tür araştırmalar için ortak bir test ortamıdır.

Bu tür bir makine öğreniminde, yapay zeka aracıları belirli bir hedefe ulaşmak için en uygun yolu bulmaya veya belirli bir görevdeki performansı artırmaya çalışır. Algoritma, hedefe doğru giden bir eylemde bulunduğunda bir ödül alır. Genel amaç: En büyük nihai ödülü kazanmak için atılacak en iyi sonraki adımı tahmin etmek.

Algoritma, seçimlerini yapmak için hem geçmiş geri bildirimlerden öğrenmeye hem de daha büyük bir getiri sağlayabilecek yeni taktikler keşfetmeye güvenir. Bu, uzun vadeli bir stratejiyi içerir - tıpkı bir satranç oyunundaki en hızlı hamle uzun vadede kazanmanıza yardımcı olmayabileceği gibi, algoritma birikimli ödülü maksimize etmeye çalışır.

# Özet
Denetimli bir öğrenme modelinde algoritma, etiketlenmiş bir veri kümesi üzerinde öğrenir ve algoritmanın eğitim verileri üzerindeki doğruluğunu değerlendirmek için kullanabileceği bir cevap anahtarı sağlar. Denetimsiz bir model ise, aksine, algoritmanın özellikleri ve örüntüleri kendi başına çıkararak anlamlandırmaya çalıştığı etiketlenmemiş veriler sağlar.
