Veri biliminde bir veri seti, genellikle satırlar ve sütunlardan oluşan iki boyutlu diziler şeklinde karşımıza çıkmaktadır. Sütunların her biri öznitelik ya da özellik (attributes - features), satırların her biri de örnekler ya da gözlemler (samples - observations) olarak adlandırılır. 

Makine öğrenmesinin başarısı, veri setinde ne kadar çok örnek bulunduğu ile doğrudan ilişkilidir. Örnek arttıkça, tahmin ve sınıflandırma doğruluğu da o kadar artacaktır.

# Etiketler (Labels)

Bir etiket, tahmin ettiğimiz şeydir. Etiket, buğdayın gelecekteki fiyatı, bir resimde gösterilen hayvan türü, bir ses dosyasının anlamı veya hemen hemen her şey olabilir. Sınıflandırma problemlerinde nihai olarak ulaşmaya çalıştığımız sınıflar da birer etikettir.

# Özellikler (Features)

Bir özellik, bir girdi değişkenidir. Basit bir makine öğrenimi projesi tek bir özelliği kullanırken, daha karmaşık bir makine öğrenimi projesi belirtilen milyonlarca özelliği kullanabilir:

$x_{1},x_{2},...,x_{N}$


Spam e-posta algılayıcı örneğinde, özellikler aşağıdakileri içerebilir:

* e-posta metnindeki kelimeler
* gönderenin adresi
* e-postanın gönderildiği günün saati

Bu modeli eğitirken, ona etiketler vermemiz gerekir. Spam e-posta filtrelemesi örneğinde, bu etiket 'spam veya spam değil' şeklinde olabilir.

Özellikler, verilerimizi temsil etme şeklimizdir. Dolayısıyla, bir e-postadan, örneğin e-postadaki kelimeler veya "gönderen adresler ve alıcı adresler", çeşitli yönlendirme veya başlık bilgileri, makine öğrenimimiz için onu temsil etmek üzere bu e-postadan çıkarabileceğimiz herhangi bir bilgi olarak düşünülebilir.

# Örnek Veriler

Bir örnek belirli bir veriden örneklenmiş bilgidir. **x** ile ifade edildiğini düşünelim.. (Bir vektör olduğunu belirtmek için x kalın harflerle yazılmıştır.) Örnekleri iki kategoriye ayırıyoruz:

* etiketlenmiş örnekler (labeled)
* etiketsiz örnekler (unlabeled)

Etiketlenmiş bir örnek, hem özellikleri hem de etiketi bir arada içerir. Yani:

`etiketlenmiş örnekler: {özellikler, etiket}: (x, y)`

Modeli eğitmek için etiketli örnekler kullanılır. Spam e-posta algılayıcı örneğimizde, etiketli örnekler kullanıcıların açıkça "spam" veya "spam değil" olarak işaretlediği tek tek e-postalar olabilir.

**Etiketsiz veri örnekleri**, doğadan alınan veya insan tarafından arkasındaki bilimsel örünütüleri keşfetmek için yaratılan verilerden oluşur. Etiketsiz verilerin bazı örnekleri arasında fotoğraflar, ses kayıtları, videolar, haber makaleleri, tweetler, röntgen filmleri vb. bulunabilir. Buradaki asıl nokta, verilerdeki özellikler için herhangi bir açıklama, etiket, sınıf veya isim olmamasıdır.

**Etiketli veri örnekleri**, verilerdeki özelliklerin açıklaması, etiketi veya adı ile ilişkilendirilmiş verilerden oluşur. Örneğin, etiketli bir görüntü veri kümesinde bir görüntü, bir kedinin fotoğrafı ve bir köpeğin fotoğrafı şeklinde etiketlenmiştir.

Meyve örneğine geri dönelim. İlk dört satırdaki veri örneklerinin hangi meyve türü olduğu bellidir. Burada meyve türü etikettir. Ancak beşinci satırdaki örneğin hangi meyveye ait olduğunu bilmemekteyiz. Bu veri örneği etiketlenmemiştir. 

>Kütle (gr) (Özellik) | Yüzey (Özellik) | Meyve Türü (Etiket)
>--- | --- | ---
>155 | Pürüzlü | Portakal
>180 | Pürüzlü | Portakal
>135 | Düz | Elma
>110 | Düz | Elma
>120 | Düz| **?**

Modelimizi etiketli örneklerle eğittikten sonra, bu modeli etiketlenmemiş örneklerdeki etiketi tahmin etmek için kullanırız. Hedef, beşinci satırdaki örneğin etiketini (yani meyve türünü) tespit edebilmektir. 

# Modeller

Bir model, özellikler ve etiket arasındaki ilişkiyi tanımlar. Örneğin, bir spam e-posta algılama modeli, belirli özellikleri "spam" ile güçlü bir şekilde ilişkilendirebilir. Bir modelin yaşamının iki aşamasını vurgulayalım:

Eğitim (training), modeli oluşturmak veya öğrenmek anlamına gelir. Yani, modele etiketli örnekleri gösterirsiniz ve modelin özellikler ile etiket arasındaki ilişkileri aşamalı olarak öğrenmesini sağlarsınız.

Çıkarım (inference), eğitimli modeli etiketlenmemiş örneklere uygulamak anlamına gelir. Yani, yararlı tahminler yapmak için eğitimli modeli kullanırsınız (y'). Örneğin, çıkarım sırasında, yeni etiketlenmemiş örnekler için meyve türünü tahmin edebilirsiniz.

# Regresyon (Bağlanım)  ya da Sınıflandırma? 

Bir regresyon modeli, sürekli değerleri tahmin etmeye çalışır. Regresyon modelleri aşağıdaki gibi sorulara cevap veren tahminlerde bulunur:

* Mersin'de bir evin değeri nedir?

* Bir kullanıcının bu reklamı tıklama olasılığı nedir?

Bir sınıflandırma modeli, ayrık değerleri tahmin eder. Sınıflandırma modelleri aşağıdaki gibi sorulara cevap veren tahminlerde bulunur:

* Verilen bir e-posta mesajı spam mı yoksa spam değil mi?

* Bu bir köpek, kedi veya hamster görüntüsü mü?

* Bu meyve bir elma mı yoksa portakal mı? 

# Veri Türleri

Veri setlerinde bulunan ve makine öğrenmesi için kullanılan veri türleri iki türe ayrılır. 

**Kategorik Veriler:** Karşılaştırılamayan ve işleme sokulamayan veri türleridir. Örneğin bir kişinin ehliyetinin olup olmaması, cinsiyeti, öğrenim durumu vb. Kategorik veriler de kendi içinde nominal ve ordinal veriler olarak ikiye ayrılır. Nominal veriler sıraya sokulamayan ve ölçülemeyen verilerdir. Nominal verilere örnek olarak bir ürünün markası verilebilir. Ordinal veriler ise sıraya sokulabilen, ancak ölçülemeyen verilerdir. Örnek olarak kapı numaraları verilebilir. 

**Sayısal Veriler:** Tamamen sayılardan oluşan, karşılaştırılabilir ve ölçülebilir verilerdir. Kişilerin yaşları ya da bir ürünün fiyatı sayısal verilere örnek olarak verilebilir. Sayısal veriler de kendi içinde oransal (ratio) ve aralık (interval) veriler olarak ikiye ayrılır. Oransal veriler, adından da anlaşılacağı üzere birbirine göre oranlanabilir veya çarpılabilir (örneğin bir cismin uzunluğu). Aralık veriler ise, çarpılamayan ve oranlanamayan sayısal verilerdir. Hava sıcaklığı verisi buna örnek olarak gösterilebilir. Sıcaklığı beş ile çarpmak herhangi bir anlam üretmeyecektir. 

Makine öğrenmesi algoritmalarının hedefi, bu veri tipleri arasında geçiş yapabilmektir. 

Kategorik veriler, sınıflandırma algoritmaları ile çalışır. Sayısal veriler ise genellikle tahmin algoritmalarında kullanılır. Sayısal veriler zamana ve tarihe bağlı olabilir (zaman serileri).
